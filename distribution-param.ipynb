## Beta Distribution Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy.stats import beta
from itertools import product
```

```python
X = np.linspace(0,1, 100)
ABs = [(0.5,0.5),(1,1),(2,2), (0.5,2), (2,0.5), (2,10)]
Y_dists = []
for items in ABs:
    A,B = items
    rv = beta(A,B)
    Y_dists.append(rv)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
leg = []
for i in range(len(ABs)):
    ax.plot(X, Y_dists[i].pdf(X))

for i in ABs:
    key = fr'{i} = $(\alpha, \beta)$'
    leg.append(key)

ax.legend(leg)

plt.savefig("images/beta-distributions/pdfs.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot()
leg = []
for i in range(len(ABs)):
    ax.plot(X, Y_dists[i].cdf(X))

for i in ABs:
    key = fr'{i} = $(\alpha, \beta)$'
    leg.append(key)

ax.legend(leg)

plt.savefig("images/beta-distributions/cdfs.png")
```
