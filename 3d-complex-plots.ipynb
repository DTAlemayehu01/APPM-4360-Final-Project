## 3D Complex Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D 
from scipy.special import gamma, zeta
from copy import deepcopy
```

## Gamma Function

```python
size = 4
count = 1000

x = np.linspace(-size, size, count)
y = np.linspace(-size, size, count)
(X,Y) = np.meshgrid(x,y)
Z1 = gamma(X + 1j*Y)
```

```python
fig = plt.figure(dpi=300)
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\Re(z)$')
ax.set_ylabel(r'$\Im(z)$')
ax.set_zlabel(r'$\Re(\Gamma(z))$')
ax.plot_surface(X,Y,Z1.real, facecolors=cm.jet(Z1.imag))

m = cm.ScalarMappable(cmap=cm.jet)
m.set_array(Z1.imag)
fig.colorbar(m, ax=ax, label=r'$\Im(\Gamma(z))$')

ax.set_box_aspect(None, zoom=0.85)

plt.savefig("images/gamma3d/gamma3d_re.png")
```

```python
fig = plt.figure(dpi=300)
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\Re(z)$')
ax.set_ylabel(r'$\Im(z)$')
ax.set_zlabel(r'$\Im(\Gamma(z))$')
ax.plot_surface(X,Y,Z1.imag,facecolors=cm.jet(Z1.real))

m = cm.ScalarMappable(cmap=cm.jet)
m.set_array(Z1.real)
fig.colorbar(m, ax=ax, label=r'$\Re(\Gamma(z))$')

ax.set_box_aspect(None, zoom=0.85)

plt.savefig("images/gamma3d/gamma3d_im.png")
```

```python
fig = plt.figure(dpi=300)
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\Re(z)$')
ax.set_ylabel(r'$\Im(z)$')
ax.set_zlabel(r'$|\Gamma(z)|$')
ax.plot_surface(X,Y,np.absolute(Z1), facecolors=cm.jet(np.angle(Z1)))

m = cm.ScalarMappable(cmap=cm.jet)
m.set_array(np.angle(Z1))
fig.colorbar(m, ax=ax, label=r'$\arg(\Gamma(z))$')

ax.set_box_aspect(None, zoom=0.85)

plt.savefig("images/gamma3d/gamma3d_mod.png")
```

```python
fig = plt.figure(dpi=300)
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\Re(z)$')
ax.set_ylabel(r'$\Im(z)$')
ax.set_zlabel(r'$|\Gamma(z)|$')
ax.set_zlim3d(0,2)
ax.plot_surface(X,Y,np.absolute(Z1),facecolors=cm.jet(np.angle(Z1)), axlim_clip=True)

m = cm.ScalarMappable(cmap=cm.jet)
m.set_array(np.angle(Z1))
fig.colorbar(m, ax=ax, label=r'$\arg(\Gamma(z))$')

ax.set_box_aspect(None, zoom=0.85)

plt.savefig("images/gamma3d/gamma3d_mod_lim.png")
```

## Reimann Zeta Function

```python
size = 20
count = 1000

x = np.linspace(-size, size, count)
y = np.linspace(-size, size, count)
(X,Y) = np.meshgrid(x,y)
Z2 = zeta(X + 1j*Y)
```
